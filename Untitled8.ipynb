```python
import pandas as pd
```

```python
df = pd.read_csv('sales_data_sample.csv', encoding='latin-1')

#Fisrt 5 rows of the dataset
print(df.head())
```

```python
# Identify and handle missing values using .isnull() in Python or filters in Excel.
df.isnull().sum()
```

```python
#Remove the columns which have max NA values
df = df.drop(columns = ['ADDRESSLINE2', 'STATE', 'TERRITORY'])
df.shape
```

```python
# Fill the na values using mean for numerical data
numerical_cols = df.select_dtypes(include=['number']).columns

# Calculate the mean for each numerical column
for col in numerical_cols:
  # Calculate the mean of non-missing values in the column
  col_mean = df[col].mean()

  # Fill missing values in the column with the calculated mean
  df[col] = df[col].fillna(col_mean)

print(df.head())
```

```python
# Fill the na values using forward fill for categorical data
categorical_cols = df.select_dtypes(include=['object']).columns

for col in categorical_cols:
  # Fill missing values in the column with the forward fill
  df[col] = df[col].ffill()

print(df.head())
```

```python
# Remove duplicate rows using .drop_duplicates() or Excel’s “Remove Duplicates”.
duplicate_rows = df[df.duplicated(keep=False)]
print("Numder of duplicates:",duplicate_rows.shape)

df = df.drop_duplicates()
print("\nDataFrame shape after removing duplicate:",df.shape)
```

```python
# Standardize text values like country names, etc.
print("Country names before standardizing")
print(df['COUNTRY'].unique())

def standardize_country(country):
    country = str(country).strip().lower()

    if country == 'usa' or country == 'u.s.a':
        return 'United States'
    elif country == 'uk':
        return 'United Kingdom'
    # Add more conditions for other country variations if needed
    else:
        return country.title()  # Capitalize first letter of each word

df["COUNTRY"] = df["COUNTRY"].apply(standardize_country)

print("\nCountry names after standardizing")
print(df['COUNTRY'].unique())
```

```python
# Convert date formats to a consistent type (e.g., dd-mm-yyyy).
print("Before standardizing:")
print(df['ORDERDATE'].head())

# Normalize separators
df['ORDERDATE'] = df['ORDERDATE'].str.replace('-', '/', regex=False)

# Convert using explicit format: MM/DD/YYYY H:M
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'], format='%m/%d/%Y %H:%M', errors='coerce')

# Format to dd-mm-yyyy
df['ORDERDATE'] = df['ORDERDATE'].dt.strftime('%d-%m-%Y')

print("\nAfter standardizing:")
df['ORDERDATE'].head()
```

```python
# Rename column headers to be clean and uniform (e.g., lowercase, no spaces).
df.columns = df.columns.str.lower().str.replace(' ', '_')
print(df.columns)
```

```python
# Check and fix data types (e.g., age should be int, date as datetime).
print("data types before changing:")
print(df.dtypes)

df['orderdate'] = pd.to_datetime(df['orderdate'],dayfirst=True, errors='coerce')

print("\ndata types after changing:")
print(df.dtypes)
```
